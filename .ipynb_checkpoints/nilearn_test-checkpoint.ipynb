```python
import numpy as np
import pandas as pd
import nibabel as nb
import nilearn as nl
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
import seaborn as sns
import matplotlib.pylab as plt

print(nl.__version__)
```

```python
im_fn = '../tia_100_out/fmriprep/sub-tia001/func/sub-tia001_task-rest_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz'
confounds_fn = '../tia_100_out/fmriprep/sub-tia001/func/sub-tia001_task-rest_desc-all_confs.csv'
```

```python
atlas = datasets.fetch_atlas_aal()
masker = NiftiLabelsMasker(labels_img = atlas['maps'], standardize = True)
time_series = masker.fit_transform(im_fn)
corr_mat = np.corrcoef(time_series.T)

fig, ax = plt.subplots(1,2, figsize = [10, 5])
mask = np.tril_indices_from(corr_mat, k = -1)
sns.heatmap(corr_mat, cmap = 'RdYlBu_r', ax = ax[0]);
sns.distplot(corr_mat[mask], ax = ax[1]);
```

```python
atlas = datasets.fetch_atlas_aal()
masker = NiftiLabelsMasker(labels_img = atlas['maps'], standardize = True)
time_series = masker.fit_transform(im_fn, confounds = confounds_fn)
corr_mat1 = np.corrcoef(time_series.T)

fig, ax = plt.subplots(1,2, figsize = [10, 5])
mask = np.tril_indices_from(corr_mat1, k = -1)
sns.heatmap(corr_mat1, cmap = 'RdYlBu_r', ax = ax[0]);
sns.distplot(corr_mat1[mask], ax = ax[1]);
```

```python
df = pd.read_csv('../tia_100_out/fmriprep/sub-tia001/func/sub-tia001_task-rest_atlas-aal_timeseries_mean.tsv', sep = '\t', index_col = 0)
corr_mat2 = df.corr().values

fig, ax = plt.subplots(1,2, figsize = [10, 5])
mask = np.tril_indices_from(corr_mat2, k = -1)
sns.heatmap(corr_mat2, cmap = 'RdYlBu_r', ax = ax[0]);
sns.distplot(corr_mat2[mask], ax = ax[1]);
```
